```python
import pandas as pd
df=pd.read_csv("mtcars.csv")
```

```python
print("The car with the best mpg")
df.nlargest(1,columns=["mpg"])
```

```python
print("The car with the worst mpg")
df.nsmallest(1,columns=["mpg"])
```

```python
print("The car with best qsec")
df.nlargest(1,columns=["qsec"])
```
